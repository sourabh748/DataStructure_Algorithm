```python
def longest_common_subsequence(seq1, seq2, idx1=0, idx2=0):
    if idx1 == len(seq1) or idx2 == len(seq2):
        return 0
    
    if seq1[idx1] == seq2[idx2]:
        return (1 + longest_common_subsequence(seq1, seq2, idx1 + 1, idx2 + 1))
    
    return max(longest_common_subsequence(seq1, seq2, idx1 + 1, idx2), 
                longest_common_subsequence(seq1, seq2, idx1, idx2 + 1))
```

```python
longest_common_subsequence("serendipitous", "precipitation")
```

```python
def get_longest_subsequence(seq1, seq2, idx1: int = 0, idx2: int= 0):
    if idx1 == len(seq1) or idx2 == len(seq2):
        return ""
    
    if seq1[idx1] == seq2[idx2]:
        return seq1[idx1] + get_longest_subsequence(seq1, seq2, idx1 + 1, idx2 + 1)
    
    leftSubSequence = get_longest_subsequence(seq1, seq2, idx1 + 1, idx2)
    rightSubSequence = get_longest_subsequence(seq1, seq2, idx1, idx2 + 1)

    if len(leftSubSequence) > len(rightSubSequence):
        return leftSubSequence
    return rightSubSequence
```

```python
get_longest_subsequence("serendipitous", "precipitation")
```

```python
def lcs_memo(seq1, seq2):
    memo = {}
    def recurse(idx1 =0, idx2 =0):
        key = (idx1, idx2)
        value = memo.get(key, -1)

        if value != -1:
            return value
        elif idx1 == len(seq1) or idx2 == len(seq2):
            memo[key] = 0
        elif seq1[idx1] == seq2[idx2]:
            memo[key] = 1 + recurse(idx1 + 1, idx2 + 1)
        else:
            memo[key] = max(recurse(idx1 + 1, idx2), recurse(idx1, idx2 + 1))
        return memo[key]
    return recurse(0, 0)
```

```python
evaluate_test_cases(lcs_memo, tescase)
```

```python
def countDistinctWays(n: int) -> int:
    if n <= 1:
        return 1
    
    return countDistinctWays( n - 1 ) + countDistinctWays( n -2 )
```
